```python
import math
from decimal import *

# реализация факториала "в лоб" по определению
def factorial(n):
    if n in [0, 1]:
        return 1
    else:
        r = 1
        for i in range(2, n+1):
            r *= i
        return r

# "быстрый" факториал с помощью оценки Стирлинга
def fast_factorial(n):
    k = Decimal(math.sqrt(math.pi * 2 * n))
    return math.ceil(Decimal(k * (Decimal(n / math.e) ** Decimal(n))))

def combined_factorial(n):
    if n < 200:
        return factorial(n)
    else:
        return fast_factorial(n)
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})

plt.figure(figsize=(10,5))
plt.plot(
        [fast_factorial(n) / factorial(n) for n in range(6, 1000)], 
        label="аппроксимация Стирлинга", 
        linewidth=4, 
        linestyle=':',
        c='black'
)
plt.plot(
        [1 for n in range(6, 1000)],
        label="n!",
        linewidth=4,
        linestyle='--',
        c='black'
)
plt.legend()
plt.grid(True)
plt.xlabel('n')
plt.ylabel('f(n)/n!')
plt.show()
```

```python
how_long_a = len(str(fast_factorial(10000)))
print("Стирлинг (цифр) = ", how_long_a)
how_long_b = len(str(factorial(10000)))
print("наивный n! (цифр) =", how_long_b)
```
