```python
import math
from decimal import *

def factorial(n):
    if n in [0, 1]:
        return 1
    else:
        r = 1
        for i in range(2, n+1):
            r *= i
        return r

def fast_factorial(n):
    k = Decimal(math.sqrt(math.pi * 2 * n))
    return math.ceil(Decimal(k * (Decimal(n / math.e) ** Decimal(n))))

def combined_factorial(n):
    if n < 200:
        return factorial(n)
    else:
        return fast_factorial(n)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot([fast_factorial(n) / factorial(n) for n in range(6, 1000)], label="аппроксимация Стирлинга", linewidth=4)
plt.plot([1 for n in range(6, 1000)],label="n!", linewidth=4)
plt.legend()
plt.grid(True)
plt.xlabel('n')
plt.ylabel('f(n)/n!')
plt.show()
```
