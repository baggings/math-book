```python
import numpy, matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
```

```python
s = ['a', 'b', 'c', 'd']         # имена
a = numpy.array([150, 15])
b = numpy.array([140, 14])
c = numpy.array([50, 9.9])
d = numpy.array([105, 10])
men = [a, b, c]         # 3 спортсмена
```

```python
# покажем спортсменов в пространстве сила-неповоротливость
plt.xlim(0, 160)
plt.ylim(0, 20)
plt.plot([x[0] for x in men], [x[1] for x in men], '*', c='black', markersize=10)
for man in range(len(men)):
    plt.annotate(s[man], xy=men[man], xytext=(-5, 10), textcoords='offset points')
plt.grid(True)
plt.xlabel('Сила рук')
plt.ylabel('Неповоротливость')
plt.show()
```

```python
# расстояние между спортсменами в ненормированном пространстве
m12 = numpy.linalg.norm(a - b)
m13 = numpy.linalg.norm(a - c)
m14 = numpy.linalg.norm(a - d)
m34 = numpy.linalg.norm(c - d)
print("|a - b| = {:.2f} << |a - c| = {:.2f}".format(m12, m13))
print("|a - d| = {:.2f} <  |c - d| = {:.2f}".format(m14, m34))
```

```python
# нормируем по каждому измерению
men = [a, b, c, d]
f = lambda x: [ (2 * x[0] - 486) / 486, (12 - x[1]) / (12 - 9.58)]
data = numpy.array([f(x) for x in men])
```

```python
plt.plot([x[0] for x in data], [x[1] for x in data], '*', markersize=10, color='black')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
for man in range(len(data)):
    plt.annotate(s[man], xy=data[man], xytext=(-5, 10), textcoords='offset points')
plt.grid(True)
plt.xlabel('Сила')
plt.ylabel('Скорость')
plt.show()
```

```python
m12 = numpy.linalg.norm(data[0] - data[1])
m13 = numpy.linalg.norm(data[0] - data[2])
m14 = numpy.linalg.norm(data[0] - data[3])
m34 = numpy.linalg.norm(data[2] - data[3])
print("|a' - b'| = {:.2f} > |c' - d'| = {:.2f}".format(m12, m34))
```

```python
records = [12, 10.8, 10.6, 10.4, 10.3, 10.2, 10.1, 10, 9.9, 9.95, 9.93, 9.92, 9.9, 9.86, 9.85, 9.84, 9.79, 9.77, 9.74, 9.72, 9.69, 9.58
]
years = [1896, 1908, 1912, 1921, 1930, 1936, 1956, 1960, 1968, 1968, 1983, 1988, 1991, 1991, 1994, 1996, 1999, 2005, 2007, 2008, 2008, 2009
]
plt.plot(years, records, 'o', markersize=10, color='black')
plt.grid(True)
plt.xlabel('год')
plt.ylabel('рекорд стометровки')
plt.show()
```

```python
# нормализация
normalized_records = numpy.log(numpy.array(records) - 9.5)
plt.plot(years, normalized_records, '*', markersize=10, color='gray')
f = lambda y: (1930 - y) / 50
plt.plot(range(1900, 2005), list(map(f, range(1900, 2005))), color='black', lw=3)
plt.grid(True)
plt.xlabel('год')
plt.ylabel('ln(рекорд - 9.5)')
plt.show()
```

```python
f = lambda x: [ 1 + (2 * x[0] - 486) / 486, 1.3 + (12 - x[1]) / (12 - 9.58) / 3]
data = numpy.log(numpy.array(list(map(f, men))))
```
